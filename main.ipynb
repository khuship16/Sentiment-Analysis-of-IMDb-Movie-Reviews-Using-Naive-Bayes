```python
from naive_bayes import *
from naive_bayes_bigram import *
from naive_bayes_tfidf import *
from naive_bayes_bigram_tfidf import *

import matplotlib.pyplot as plt
```

```python
unigram_nb = NaiveBayes()
unigram_nb.train_naive_bayes()
unigram_nb.naive_bayes_test_df()
uni_acc, uni_prec, uni_rec = unigram_nb.print_statistics()
```

```python
bigram_nb = BigramNaiveBayes()
bigram_nb.train_naive_bayes()
bigram_nb.naive_bayes_test_df()
bi_acc, bi_prec, bi_rec = bigram_nb.print_statistics()
```

```python
tfidf_nb = NaiveBayes_IDF()
tfidf_nb.train_naive_bayes()
tfidf_nb.naive_bayes_test_df()
idf_acc, idf_prec, idf_rec = tfidf_nb.print_statistics()
```

```python
mixture_nb = BigramNaiveBayes_IDF()
mixture_nb.train_naive_bayes()
mixture_nb.naive_bayes_test_df()
mix_acc, mix_prec, mix_rec = mixture_nb.print_statistics()
```

```python
x_1 = ["Unigram", "IDF", "Bigram", "Mixture"]
values_1 = [uni_acc, idf_acc, bi_acc, mix_acc]

plt.bar(x_1, values_1)
plt.ylim(0.855, 0.89)
plt.ylabel("Accuracy")
plt.xlabel("Model")
plt.title("Accuracy over Naive Bayes Models")
plt.show()
```

```python
x_2 = ["Unigram", "IDF", "Bigram", "Mixture"]
values_2 = [uni_prec, idf_prec, bi_prec, mix_prec]

plt.bar(x_2, values_2)
plt.ylim(0.86, 0.91)
plt.ylabel("Precision")
plt.xlabel("Model")
plt.title("Precision over Naive Bayes Models")
plt.show()
```

```python
x_3 = ["Unigram", "IDF", "Bigram", "Mixture"]
values_3 = [uni_rec, idf_rec, bi_rec, mix_rec]

plt.bar(x_3, values_3)
plt.ylim(0.83, 0.875)
plt.ylabel("Recall")
plt.xlabel("Model")
plt.title("Recall over Naive Bayes Models")
plt.show()
```
